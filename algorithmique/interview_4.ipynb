# 1 tradeof entre donnes en mmoire et nombre d'oprations : plus de donnes mmoire possibles / rduction du nombre d'opration



```python
def intToRoman( num):
        ones = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]
        tens = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
        hrns = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
        ths = ["", "M", "MM", "MMM"]

        return ths[num // 1000] + hrns[(num % 1000) // 100] + tens[(num % 100) // 10] + ones[num % 10]
```

# 2) ne stocker que les cas  difficiles

```python
def intToRoman( num: int) -> str:
        value_to_symbol = {
            1000: 'M', 900: 'CM', 500: 'D', 400: 'CD',
            100: 'C', 90: 'XC', 50: 'L', 40: 'XL',
            10: 'X', 9: 'IX', 5: 'V', 4: 'IV', 1: 'I'
        }

        roman = []
        for value in sorted(value_to_symbol.keys(), reverse=True):

            while num >= value:
                roman.append(value_to_symbol[value])
                num -= value

        return ''.join(roman)
```
